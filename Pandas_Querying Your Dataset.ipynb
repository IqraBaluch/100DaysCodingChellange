# Querying Your Dataset

```python
import requests

download_url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-elo/nbaallelo.csv"
target_csv_path = "nba_all_elo.csv"

response = requests.get(download_url)
response.raise_for_status()    # Check that the request was successful
with open(target_csv_path, "wb") as f:
    f.write(response.content)
print("Download ready.")
```

```python
import pandas as pd
nba = pd.read_csv("nba_all_elo.csv")
nba.head()
```

 select rows based on the values in your dataset’s columns to query your data.

# select the rows where a specific field is not null. 
We have NaN in the column of 'notes'<br>
This can be helpful,to avoid any missing values in a column. 

```python
games_with_notes = nba[nba['notes'].notnull()]
```

```python
games_with_notes.head(10)
```

.notna() also be used to achieve the same goal.(removing NaN rows)

```python
games_without_null = nba[nba['notes'].notna()]
```

```python
games_without_null.head()
```

# access values of the object data type as str and perform string methods 

```python
# use .str.endswith() to filter your dataset and find all games where the home team’s name ends with "ers".
ers = nba[nba['fran_id'].str.endswith('ers')]
```

```python
ers.head()
```

# combine multiple criteria and query of your dataset as well. To do this, be sure to put each one in parentheses and use the logical operators | and & to separate them.

```python
combination = nba[

    (nba['team_id'] == 'BLB') &
    (nba['_iscopy'] == 0) &
    (nba['pts'] > 100 )&
    (nba['opp_pts'] > 100)
]
```

```python
combination.head()
```

In the spring of 1992, both teams from Los Angeles had to play a home game at another court. Query your dataset to find those two games. Both teams have an ID starting with "LA". 

```python
nba[
    (nba['_iscopy']== 0)&
    (nba['year_id'] == 1992) &
    (nba['game_id'].str.startswith('LA'))&
    (nba['notes'].notnull())
]
```
