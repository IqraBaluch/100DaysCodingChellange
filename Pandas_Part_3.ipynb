# Using .loc and .iloc
A DataFrame also provides .loc and .iloc data access methods.
### Working on rows of DataFrame
1. TO access all the data from first rows of every column, where as indexing starts from 0 

```python

data.loc[0]
```

2. TO access all the data from second rows of every column, where as indexing starts from 0 there for .loc[1] means second rows

```python
data.loc[1]
```

accessing DataFrame elements with dot notation doesnot work if column name is accidently any attribute's or methods name (.shape)
# .shape attribute
.shape attribute is used to see the number of rows and columns in a DataFrame. 

```python
data1 = pd.DataFrame([
    {'name' : "laptop", 'shape' : 'rectangle'},
    {'name' : 'mobile', 'shape' : 'cube'}
    
],index = ['one', 'two'])
```

```python
data1['shape']
```

```python
data1.shape
```

```python
# Select row with label/Index one
data1.loc['one']
```

```python
# Select row with label two
data1.loc['two']
```

```python
# Select row at index 1
data1.iloc[1]
```

```python
# Selects data at index 0
data1.iloc[0]
```

```python
# Lock data to shape and just access data of shapes
data1.loc['one':'two', 'shape']
```

```python
# Lock data to name and just access data of name
data1.loc['one': 'two','name']
```
