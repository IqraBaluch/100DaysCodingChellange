# merge dataFrame with itself

```python
import pandas as pd
```

```python
data = pd.read_csv('Desktop/netflix_titles.csv')
```

```python
data.head()
```

# Merge with id

```python
mer = data.merge(data, on= 'show_id', how = 'inner', suffixes =('_er', '_ed'))
```

```python
mer.head()
```

# Create a boolean index to select the appropriate rows

```python
boolean_filter = ((mer['country_er']=='United States'))
                  

us_er = mer[boolean_filter]
```

```python
us_er
```

```python
boolean_filter1 = ((mer['country_er']!='United States') | (mer['country_ed']=='Brazil'))
                  

us_er_2 = mer[boolean_filter1]
```
